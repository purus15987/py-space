#Handling Outliers

##Finding Outliers in Gene1 Set

```python
TPM_df = P1_TPM_df.copy()
TPM_df_genes = TPM_df.iloc[:, 1:]
gene1_name = TPM_df.iloc[0,0]

# get data for firt gene
TPM_df_gene1 = TPM_df_genes.iloc[0,:]
print('orininal shape : ',TPM_df_gene1.shape)
# Identify and handle outliers
q1 = np.percentile(TPM_df_gene1, 25)
q3 = np.percentile(TPM_df_gene1, 75)
iqr = q3 - q1
upper_limit = q3 + 1.5 * iqr
lower_limit = q1 - 1.5 * iqr
df = TPM_df_gene1[(TPM_df_gene1 > lower_limit) & (TPM_df_gene1 < upper_limit)]
print('data shape after Outliers Removed : ',df.shape)
print(df)
```

```python
import matplotlib.pyplot as plt
# Plot the data
plt.figure(figsize=(10, 6))

# Original data
plt.subplot(2, 1, 1)
plt.boxplot(TPM_df_gene1, vert=False)
plt.title('Original Data for Gene 1 :'+gene1_name)

# Data with outliers removed
plt.subplot(2, 1, 2)
plt.boxplot(df, vert=False)
plt.title('Data with Outliers Removed for Gene 1 : '+gene1_name)

plt.tight_layout()
plt.show()
```

##Finding Outliers in ALL Genes in P1 Dataset

```python
import pandas as pd

TPM_df1 = P1_TPM_df.copy()

TPM_df_genes = TPM_df1.iloc[:, 1:]

# Calculate the IQR for each column
Q1 = TPM_df_genes.quantile(0.25)
Q3 = TPM_df_genes.quantile(0.75)
IQR = Q3 - Q1

# Set an IQR threshold (e.g., 1.5 times the IQR)
threshold = 1.5

# Identify outliers based on the threshold
outliers1 = ((TPM_df_genes < (Q1 - threshold * IQR)) | (TPM_df_genes > (Q3 + threshold * IQR)))

# print(outliers)

# Display rows and columns containing outliers1
outliers_df1 = TPM_df[outliers1.any(axis=1)]
print("Rows with outliers:",outliers_df1.shape[0])
outliers_df1.head()
```

## Finding Outliers in ALL Genes in P2 Dataset

```python
import pandas as pd

TPM_df2 = P2_TPM_df.copy()

TPM_df_genes = TPM_df2.iloc[:, 1:]

# Calculate the IQR for each column
Q1 = TPM_df_genes.quantile(0.25)
Q3 = TPM_df_genes.quantile(0.75)
IQR = Q3 - Q1

# Set an IQR threshold (e.g., 1.5 times the IQR)
threshold = 1.5

# Identify outliers based on the threshold
outliers2 = ((TPM_df_genes < (Q1 - threshold * IQR)) | (TPM_df_genes > (Q3 + threshold * IQR)))

# print(outliers)

# Display rows and columns containing outliers2
outliers_df2 = TPM_df[outliers2.any(axis=1)]
print("Rows with outliers:",outliers_df2.shape[0])
outliers_df2.head()
```
