# Image to Text

```python
from PIL import Image
import pytesseract

pytesseract.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'

# Perform OCR on an image file
def extract_text_from_image(image_path):
    img = Image.open(image_path)

    # Use Tesseract to do OCR on the image
    text = pytesseract.image_to_string(img, lang='eng', config='--psm 6')  # Adjust config if needed

    return text

# Specify the path to your image file
image_path = '//content/Gate Scorecard.jpeg'  # Replace with the actual path to your image

# Perform OCR on the image
result_text = extract_text_from_image(image_path)

# Print the extracted text
print("Extracted Text:")
print(result_text)
```

# Extract text from images in pdf

```python
import fitz  # PyMuPDF
from PIL import Image
import pytesseract
import io
from docx import Document
from docx.shared import Pt

# Set the Tesseract executable path manually (default location in Colab)
pytesseract.pytesseract.tesseract_cmd = '/usr/bin/tesseract'

def clean_text(text):
    # Remove non-printable characters except newlines
    printable_text = ''.join(char if char.isprintable() or char == '\n' else ' ' for char in text)
    return printable_text

def extract_text_from_pdf(pdf_path):
    # Open the PDF file
    pdf_document = fitz.open(pdf_path)

    # Initialize an empty list to store the extracted text
    text_list = []

    print('no. of pages:',pdf_document.page_count)

    # Iterate through each page in the PDF
    for page_num in range(pdf_document.page_count):
        print('extracting page :', page_num+1)
        # Get the page
        page = pdf_document[page_num]

        # Get the pixmap of the page (higher resolution image)
        pixmap = page.get_pixmap()

        # Convert the pixmap to a PIL Image
        image = Image.frombytes("RGB", [pixmap.width, pixmap.height], pixmap.samples)

        # Use Tesseract to do OCR on the image
        image_text = pytesseract.image_to_string(image, lang='eng', config='--psm 6')
        cleaned_text = clean_text(image_text)
        # print(cleaned_text)

        # Append the extracted text to the list
        text_list.append(f"Page {page_num + 1}:\n{cleaned_text}\n\n\n")

    # Close the PDF document
    pdf_document.close()

    return text_list

def create_word_document(text_list, output_path):
    # Create a new Word document
    doc = Document()

    # Set font size for the document
    style = doc.styles['Normal']
    font = style.font
    font.name = 'Arial'
    font.size = Pt(12)

    # Add text to the Word document
    for text in text_list:
        doc.add_paragraph(text)

    # Save the Word document
    doc.save(output_path)

# Replace 'your_pdf_path.pdf' with the path to your PDF file in Colab
pdf_path = '/content/6th chapter .pdf'
output_path = '/content/6th chapter.docx'

# Extract text from the PDF
result_text_list = extract_text_from_pdf(pdf_path)

# Create and save the Word document
create_word_document(result_text_list, output_path)

print("Word document saved successfully.")
```
