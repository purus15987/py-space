```python
import pandas as pd
from langdetect import detect
from googletrans import Translator
from tqdm import tqdm

tqdm.pandas()  # Enable tqdm for Pandas

# Initialize the translator
translator = Translator()

# Function to detect language and translate if necessary
def translate_content(content):
    try:
        # Detect the language of the content
        detected_lang = detect(content)

        # If the content is not in English, translate it
        if detected_lang != 'en':
            translated = translator.translate(content, src=detected_lang, dest='en')
            return translated.text
        else:
            return content
    except Exception as e:
        return content  # Return the original content if an error occurs

# Apply the function to the 'content' column
df.loc[:19,'translated_tweet'] = df.loc[:19,'tweet'].progress_apply(translate_content)
```

# Data Cleaning

```python
import nltk

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.probability import FreqDist
from tqdm import tqdm


# Enable tqdm for Pandas
tqdm.pandas()

# Define the function for cleaning the text data
def clean_Content(text, tweetUrl):
    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # Remove non-alphanumeric characters (except spaces)
    text = re.sub(r'[^a-z0-9A-Z\s#]', '', text)

    # Tokenization: Split text into words
    tokens = text.split()

    # Handle split of words like #StopHateAgainstFarmers into ['Stop', 'Hate', 'Against', 'Farmers']
    processed_tokens = []
    for token in tokens:
        if token.startswith('#'):  # If the word starts with #
            # Split the word by uppercase letters (CamelCase)
            token_split = re.sub('([a-z0-9])([A-Z])', r'\1 \2', token[1:])  # Skip the '#'
            processed_tokens.extend(token_split.split())
        else:
            processed_tokens.append(token)

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in processed_tokens if word not in stop_words]

    # Stemming: Apply Porter Stemmer
    stemmer = PorterStemmer()
    tokens_stemmed = [stemmer.stem(word) for word in tokens]

    # Lemmatization: Apply WordNet Lemmatizer
    lemmatizer = WordNetLemmatizer()
    tokens_lemmatized = [lemmatizer.lemmatize(word) for word in tokens_stemmed]  # Lemmatize after stemming

    # Join the tokens back into a cleaned string
    cleaned_text = ' '.join(tokens_lemmatized)

    # extract user_name
    user_name = re.search(r'twitter.com/([^/]+)/status/', tweetUrl)
    if user_name:
        user_name = ' user' + user_name.group(1)  # Return the username
    else:
        user_name = ' user'


    # Append user_name
    cleaned_text += user_name

    # Convert to lowercase
    cleaned_text = cleaned_text.lower()

    return cleaned_text

# Example: Apply the cleaning function to the 'tweet' column of your DataFrame
df['cleanedContent'] = df.progress_apply(
    lambda row: clean_Content(str(row['tweet']), row['link']),
    axis=1
)
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from tqdm import tqdm


# Enable tqdm for Pandas
tqdm.pandas()

# Initialize VADER sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Define analyze sentiment
def analyze_sentiment(text):
    sentiment_score = sia.polarity_scores(text)
    return sentiment_score['compound']  # Compound score gives an overall sentiment

# Apply sentiment analysis to the 'renderedContent' column
df['sentiment-analysis'] = df['cleanedContent'].progress_apply(analyze_sentiment)

# Classify as misinformation (negative sentiment) or non-misinformation (positive sentiment)
# You can adjust the threshold based on your dataset and definition of misinformation
df['information'] = df['sentiment-analysis'].progress_apply(lambda x: 'negitive' if x < 0.05 else 'positive')

# save results to 'results.csv'
df.to_csv('results.csv')
```

```python
import torch
from transformers import DistilBertForSequenceClassification, Trainer, TrainingArguments, DistilBertTokenizer
from sklearn.model_selection import train_test_split
import pandas as pd
from torch.utils.data import DataLoader

# Check if GPU is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Load the dataset
df = pd.read_csv('bert_results.csv')

# Prepare your dataset (assuming you have a 'cleanedContent' column and a 'information' column)
data = df[['cleanedContent', 'information']]

# Split your data into training and validation sets
train_texts, val_texts, train_labels, val_labels = train_test_split(
    data['cleanedContent'].tolist(),
    data['information'].tolist(),
    test_size=0.2,
    random_state=42
)

# Initialize the tokenizer
tokenizer = DistilBertTokenizer.from_pretrained("distilbert-base-uncased")

# Tokenize the data
train_encodings = tokenizer(train_texts, padding=True, truncation=True, max_length=512)
val_encodings = tokenizer(val_texts, padding=True, truncation=True, max_length=512)

# Convert labels to numerical values (e.g., 0 for negative, 1 for positive)
train_labels = [1 if label == 'positive' else 0 for label in train_labels]
val_labels = [1 if label == 'positive' else 0 for label in val_labels]  # Fixed typo here

# Create torch Dataset objects
class SentimentDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}  # Removed .to(device)
        item['labels'] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)

# Create the dataset objects
train_dataset = SentimentDataset(train_encodings, train_labels)
val_dataset = SentimentDataset(val_encodings, val_labels)

# Load the model and move to GPU
model = DistilBertForSequenceClassification.from_pretrained("distilbert-base-uncased", num_labels=2).to(device)

# Set up training arguments
training_args = TrainingArguments(
    output_dir="./results",
    num_train_epochs=3,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    logging_dir="./logs",
    eval_strategy="epoch",  # Updated to eval_strategy
    save_strategy="epoch",  # Save model after every epoch
    load_best_model_at_end=True,  # Load the best model after training,
    fp16=True,  # Enable mixed precision training
    gradient_accumulation_steps=4,
)

# Initialize Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
)

# Train the model
trainer.train()

# Save the model and tokenizer
model.save_pretrained("./sentiment_model")
tokenizer.save_pretrained("./sentiment_model")
```

```python
# Evaluate the model on the validation set
eval_results = trainer.evaluate()

# Display the evaluation results
print(eval_results)
```

```python
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

# Predict the sentiment on the validation set
predictions = trainer.predict(val_dataset)

# Get the predicted labels (0 for negative, 1 for positive)
pred_labels = predictions.predictions.argmax(axis=-1)

# True labels from the validation set
true_labels = val_labels

# Compute accuracy
accuracy = accuracy_score(true_labels, pred_labels)

# Compute precision, recall, and F1-score
precision, recall, f1, _ = precision_recall_fscore_support(true_labels, pred_labels, average='binary')

# Print the metrics
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
```
